```python
import pandas as pd
```

```python
interactions_df = pd.read_csv('data/notifications.csv')
interactions_df.drop(labels=['parent_comment_id', 'status', 'created_at', 'updated_at'], axis=1, inplace=True)
```

```python
li = ['id', 'post_id', 'comment', 'sender_id', 'user_id', 'action', 'type']
for item in li:
    interactions_df[item] = interactions_df[item].fillna('')
user_list = interactions_df['user_id'].tolist()
user_list = list(dict.fromkeys(user_list))
```

```python
interactions_user_df = interactions_df.groupby(['user_id'])
```

```python
user_read = {}
for item in user_list:
    item = int(item)
    item_list=[]
    int_item_list=[]
    for i in range(len(interactions_user_df.get_group(item)['post_id'])):
        item_list.append(interactions_user_df.get_group(item)['post_id'].iloc[i])
        item_list = list(dict.fromkeys(item_list))
        if '' in item_list:
            item_list.remove('')
    for i in item_list:
        int_item_list.append(int(i))
    user_read.update({item:int_item_list})
        
```

```python
print(user_read)
```

```python
post_df = pd.read_csv('data/posts.csv')
```

```python
post_list = post_df['id'].tolist()
print(post_list)
```

```python
user_read[7]
```

```python
user_not_read = {}
for user in user_list:
    user_notread=[]
    for post in post_list:
        if post not in user_read[user]:
            user_notread.append(post)
    user_not_read.update({user:user_notread})            
```

```python
print(user_not_read)
```

```python
user_not_read_df = pd.DataFrame(list(user_not_read.items()), index = [i for i in range(len(user_not_read))], columns=['user_id', 'post_id'])
user_not_read_df.head()
```

```python
user_not_read_df.to_csv(r'data/user_not_read.csv', index=None, header=True)
```
